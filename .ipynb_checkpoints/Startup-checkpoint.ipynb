## –†–∞–∑–¥–µ–ª 2. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

### 1. –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
def na_search (dataframe):
    print(f'–í –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ {[x for x in globals() if globals()[x] is dataframe][0]}')
    count = 0
    for i in dataframe.columns:
        if dataframe[i].isna().sum() != 0:
            print(f'–∫–æ–ª-–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç–æ–ª–±—Üe {i} - {dataframe[i].isna().sum()}')    
            count +=1
    if count == 0:
        print('–ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–µ—Ç')
    else:
        print('–≤ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö –Ω–µ—Ç')
    print()
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–≤–µ–¥–µ–Ω–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –∏ —É–¥–∞–ª–µ–Ω–∏—è –ø—Ä–æ–±–µ–ª–æ–≤ –≤ –Ω–∞—á–∞–ª–µ –∏ –∫–æ–Ω—Ü–µ
def snake_cols(dataframe):
    dataframe.columns = [re.sub(r'(?<!^)(?=[A-Z])','_',i).replace(' ','_').lower().strip() for i in dataframe.columns]
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è —É–¥–∞–ª–µ–Ω–∏—è –ø—Ä–æ–±–µ–ª–æ–≤ –≤ –Ω–∞—á–∞–ª–µ –∏ –∫–æ–Ω—Ü–µ —è—á–µ–µ–∫
def snake_text(dataframe):
    for i in dataframe.columns:
        if dataframe[i].dtype == 'object':
            dataframe[i] = dataframe[i].str.replace(' ','_').str.strip()
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö
def unique_textcolumns (dataframe):
    for i in dataframe.columns:
        if dataframe[i].dtype == 'object':
            print(dataframe[i].value_counts())
            print()
```

```python
# –∑–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ –∑–∞–∫—Ä—ã—Ç–∏—è —Å—Ç–∞—Ä—Ç–∞–ø–∞ –Ω–∞ –¥–∞—Ç—É –≤—ã–≥—Ä—É–∑–∫–∏ –¥–∞—Ç–∞—Å–µ—Ç–∞
startups_train['closed_at'] = startups_train['closed_at'].fillna('2018-01-01')
# –∑–∞–º–µ–Ω—è–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö —Å –¥–∞—Ç–∞–º–∏ –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π
for i in ['founded_at','first_funding_at','last_funding_at','closed_at']:
    startups_train[i] = pd.to_datetime(startups_train[i], format='%Y-%m-%d')
    # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞—Ç—ã –≤ —á–∏—Å–ª–æ –¥–Ω–µ–π (—Å–∫–æ–ª—å–∫–æ –¥–Ω–µ–π –Ω–∞–∑–∞–¥ –æ—Ç 2018-01-01 –ø—Ä–æ–∏–∑–æ—à–ª–æ —Å–æ–±—ã—Ç–∏–µ)
    startups_train[i] = (startups_train[i] - pd.to_datetime('2018-01-01', format='%Y-%m-%d')).dt.days
```

```python
# –ø—Ä–æ–≤–æ–¥–∏–º –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫—É —Å –ø–æ–º–æ—â—å—é —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π 
snake_cols(startups_train)
```

```python
# –≤—ã–≤–æ–¥–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö
unique_textcolumns(startups_train)
```

### 1. –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π

```python
# —Å–æ–∑–¥–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤ –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
def stat_plot(dataframe):
    k = 0
    cols = dataframe.select_dtypes(include='number').columns.tolist()
    n = len(cols)
    plt.figure(figsize=(15, 6*n))
    for i in cols:
        k += 1
        plt.subplot(n,2,k)
        dataframe[i].plot(kind='hist', bins=50)
        plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ "' + i +'"')
        plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ "' + i +'"')
        plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞ –≤—Å—Ç—Ä–µ—á–∞–µ–º–æ—Å—Ç–∏')
        k += 1
        plt.subplot(n,2,k)
        dataframe[i].plot(kind='box')
        plt.title('–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–∑–º–∞—Ö–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–∞ "' + i +'"')
        plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ "' + i +'"')        
    plt.show()
```

```python
# –≤—ã–≤–æ–¥ –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ startups_train
startups_train.describe()
```

## –†–∞–∑–¥–µ–ª 5. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø–∞–π–ø–ª–∞–π–Ω–æ–≤

```python
# —Ä–∞–∑–±–∏–≤–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_valid, y_train, y_valid = train_test_split(
    startups_train.drop(['name', 'city', 'state_code', 'region', 'first_funding_at','status'], axis=1),
    startups_train['status'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = startups_train['status']
)

# —Å–æ–∑–¥–∞–µ–º —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ LabelEncoder –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
label_encoder = LabelEncoder()

# —Ñ–∏–∫—Å–∏—Ä—É–µ–º –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ y_train –¥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
y_train_first = y_train[y_train.index[0]]

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –∏ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –≤—ã–±–æ—Ä–∫—É 
y_train = label_encoder.fit_transform(y_train)

# —Ñ–∏–∫—Å–∏—Ä—É–µ–º –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ y_valid –¥–æ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
y_valid_first = y_valid[y_valid.index[0]]

# —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É
y_valid = label_encoder.transform(y_valid)

# –ø—Ä–æ–≤–µ—Ä—è–µ–º –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
print(f'–ó–Ω–∞—á–µ–Ω–∏—è "{y_train_first}" –≤ y_train –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω—ã —á–∏—Å–ª–æ–º {y_train[0]}')
print(f'–ó–Ω–∞—á–µ–Ω–∏—è "{y_valid_first}" –≤ y_valid –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω—ã —á–∏—Å–ª–æ–º {y_valid[0]}')
```

```python
# —Å–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–∫–∏ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
ohe_columns = X_train.select_dtypes(exclude='number').columns
num_columns = X_train.select_dtypes(include='number').columns

# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns: –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(strategy='constant', fill_value='unknown')),
     ('ohe', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
    ]
    )

# —Å–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(45,50),
        'models__max_features': range(35,40),
        'preprocessor__num': [StandardScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(4,6), 
        'preprocessor__num': [StandardScaler(),  'passthrough']   
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    #{
    #    'models': [LogisticRegression(
     #       random_state=RANDOM_STATE, 
    #        solver='liblinear', 
     #       penalty='l1'
    #    )],
     #   'models__C': range(1,4),
     #   'preprocessor__num': [StandardScaler(),  'passthrough']        
    #},
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='poly')],
        'models__C': [3, 4, 5],        
        'preprocessor__num': [StandardScaler(),  'passthrough']  
    },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ GaussianNB
   # {
   #     'models': [GaussianNB()],                
   #     'preprocessor__num': [StandardScaler(),  'passthrough']  
   # },
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ RandomForestClassifier
   # {
   #     'models': [RandomForestClassifier()],
   #     'models__n_estimators':[50, 55],
   #     'models__max_depth': [20, 25],        
   #     'preprocessor__num': [StandardScaler(), 'passthrough']  
   # }
]

grid = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5, 
    scoring='f1',
    n_jobs=-1,
    random_state=RANDOM_STATE,
    n_iter=50
)
grid.fit(X_train, y_train)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:')
display(grid.best_estimator_.named_steps['models'])

print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', grid.best_score_.round(4))
```

```python
# –ø—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞–±–æ—Ç—É –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
y_valid_pred = grid.predict(X_valid)
print(f'–ú–µ—Ç—Ä–∏–∫–∞ f1 –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ: {f1_score(y_valid, y_valid_pred).round(4)}')
```

```python
# —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
X_test = startups_test.drop(['name', 'city', 'region', 'state_code', 'first_funding_at'], axis=1)
y_pred = grid.predict(X_test)
```

```python
# —Å–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º–∏
Pred = pd.DataFrame({'status': y_pred}, index=startups_test['name'])
```

```python
# –∑–∞–º–µ–Ω—è–µ–º –±—É–ª–µ–≤—ã –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å—Ç—Ä–æ–∫–∏ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π –ø–µ—Ä–µ–¥–∞—á–∏ –≤ —Ñ–∞–π–ª –¥–ª—è –∫–µ–≥–ª–∏
Pred['status'] = Pred['status'].replace(1, 'operating')
Pred['status'] = Pred['status'].replace(0, 'closed')
```

–î–ª—è —ç—Ç–æ–≥–æ —Ç—Ä–µ–±–æ–≤–∞–ª–æ—Å—å —Ä–µ—à–∏—Ç—å –¥–≤–µ –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∞–Ω–Ω—ã–µ –∑–∞–¥–∞—á–∏: **—Ä–∞–∑—Ä–∞–±–æ—Ç–∞—Ç—å –º–æ–¥–µ–ª—å –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏—è –¥–µ—è—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å—Ç–∞—Ä—Ç–∞–ø–∞**, –ø—Ä–æ–≤–µ—Å—Ç–∏ –ø–æ–ª–Ω–æ—Ü–µ–Ω–Ω—ã–π —Ä–∞–∑–≤–µ–¥–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∏ **—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –±—É–¥—É—â–∏–º —Å–æ–∑–¥–∞—Ç–µ–ª—è–º —Å—Ç–∞—Ä—Ç–∞–ø–æ–≤** (–∫–∞–∫–∏–µ —Ñ–∞–∫—Ç–æ—Ä—ã –≤–ª–∏—è—é—Ç –Ω–∞ —É—Å–ø–µ—à–Ω–æ—Å—Ç—å —Å—Ç–∞—Ä—Ç–∞–ø–∞).
